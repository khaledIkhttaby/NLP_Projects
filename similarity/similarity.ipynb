# TF_IDF 

```python
model_TF=pickle.load(open("tfidf1.pkl", 'rb'))
print(model_TF)
```

```python
def clean(s):
    return [w.strip(',."!?:;()\'') for w in s]
def preprocess_word2vec(word):
    return word.split(" ")
def cosine_distance_wordembedding_method_word2vec(s1, s2):
   
    vector_1 = np.mean([model_word2vec.wv[word] for word in preprocess_word2vec(s1)],axis=0)
    vector_2 = np.mean([model_word2vec.wv[word] for word in preprocess_word2vec(s2)],axis=0)
    cosine = scipy.spatial.distance.cosine(vector_1, vector_2)
    #print('Word Embedding method with a cosine distance asses that our two sentences are similar to',round((1-cosine)*100,2),'%')
    return round((1-cosine)*100,2)
def preprocess_doc2vec(word):
    return word.split(" ")
def cosine_distance_wordembedding_method_doc2vec(vector_1, vector_2):
    vector_1=model_doc2vec.infer_vector(vector_1)
    vector_2=model_doc2vec.infer_vector(vector_2)
    cosine = scipy.spatial.distance.cosine(vector_1, vector_2)
    #print('Word Embedding method with a cosine distance asses that our two sentences are similar to',round((1-cosine)*100,2),'%')
    return round((1-cosine)*100,2)
def allsimilar_doc2vec(sentnce,data):
    dic={}
    data=data.astype(str)
    sentnce=preprocess_doc2vec(sentnce)
    data=pd.Series(data)
    #result=data.apply(lambda x:cosine_distance_wordembedding_method(sentnce,preprocess(x)))
    for i in range(len(data)):
        Sentence=preprocess_doc2vec(data[i])
        result=cosine_distance_wordembedding_method_doc2vec(sentnce,Sentence)
        #print(i)
        dic[i]=result
    return dic
def getindexsentrnce_doc2vec(dicsimilar,number_of_most):
    s=0
    indexsentence=[]
    for k in sorted(dicsimilar,key=dicsimilar.get,reverse=True):
        if(s>number_of_most):
            break
        indexsentence.append(k)
        s+=1
    return indexsentence
def allsimilar_word2vec(sentnce,data):
    dic={}
    data=data.astype(str)
    for i in range(len(data)):
        result=cosine_distance_wordembedding_method_word2vec(sentnce,data[i])
        dic[i]=result
    return dic
def getindexsentrnce_word2vec(dicsimilar,number_of_most):
    s=0
    indexsentence=[]
    for k in sorted(dicsimilar,key=dicsimilar.get,reverse=True):
        
        if(s>number_of_most):
            break
        indexsentence.append(k)
        s+=1
    return indexsentence
def getvectorfromsntence(sentence):
    sentence=[sentence]
    return model_TF.transform(sentence)
    
def cosine_distance_wordembedding_method_tf(s1,s2):
    vector_1=getvectorfromsntence(s1)
    vector_2=getvectorfromsntence(s2)
    cosine = scipy.spatial.distance.cosine(vector_1[0].todense(), vector_2[0].todense())
    #print('Word Embedding method with a cosine distance asses that our two sentences are similar to',round((1-cosine)*100,2),'%')
    return round((1-cosine)*100,2)
def allsimilar_tf(sentnce,data):
    dic={}
    data=data.astype(str)
    for i in range(len(data)):
        result=cosine_distance_wordembedding_method_tf(sentnce,data[i])
        dic[i]=result
    return dic
def getindexsentrnce_tf(dicsimilar,number_of_most):
    s=0
    indexsentence=[]
    for k in sorted(dicsimilar,key=dicsimilar.get,reverse=True):
        
        if(s>number_of_most):
            break
        indexsentence.append(k)
        s+=1
    return indexsentence
```

```python
s2_word2vec="ميسي برشلونة"
dic_word2vec=allsimilar_word2vec(s2_word2vec,Data)
indexstatments_word2vec=getindexsentrnce_word2vec(dic_word2vec,20)
statment_word2vec=pd.DataFrame([Data[i] for i in indexstatments_word2vec],[dic_word2vec[i] for i in indexstatments_word2vec],[s2_word2vec])
statment_word2vec
```

```python
statment_word2vec.to_csv("similarity_word2vec.csv")
```

```python
s2_doc2vec="ميسي برشلونة"
dic_doc2vec=allsimilar_doc2vec(s2_doc2vec,Data)
indexstatments_doc2vec=getindexsentrnce_doc2vec(dic_doc2vec,20)
statment_doc2vec=pd.DataFrame([Data[i] for i in indexstatments_doc2vec],[dic_doc2vec[i] for i in indexstatments_doc2vec],[s2_doc2vec]) 
statment_doc2vec
```

```python
s2_tf="ميسي برشلونة"
dic_tf=allsimilar_tf(s2_tf,Data)
indexstatments_tf=getindexsentrnce_tf(dic_tf,20)
statment_tf=pd.DataFrame([Data[i] for i in indexstatments_tf],[dic_tf[i] for i in indexstatments_tf],[s2_tf])
statment_tf
```
