```python
def generatwords(word):
    length=len(word)
    result=[]
    currentword=''
    return genwor(word,length,currentword,0,result)
```

```python

def genwor(word,length,currentword,i,r):
    if(i<length):
        for x in range(length):
            if((currentword.count(word[x]))<1 and len(currentword)<2 or ((currentword.count(word[x]))==1 and len(currentword)>=2  )):
                currentword=currentword+word[x]
                if(len(currentword)>1):
                    r.append(currentword)
                i+=1
                genwor(word,length,currentword,i,r)
                currentword=currentword.replace(word[x],'')
                i-=1
        return r  
```
