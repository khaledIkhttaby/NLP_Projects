```python
def preprocessing_post(post):
    import re
    def delete_english_char(post):
        return re.sub(r"[A-Za-z]"," ",post)
    def normaalize_write(post):
        post = re.sub("[إأٱآا]", "ا", post)
        post = re.sub("ى", "ي", post)
        post = re.sub("ؤ", "ء", post)
        post = re.sub("[ﻷ]", "لا", post)
        post = re.sub("ة", "ه", post)
#         post = re.sub("ئ", "ء", post)
#         post = ''.join([i for i in post if not i.isdigit()])
        post = re.sub(r'([\u0600-\u06FF])\1{2,}', r'\1', post)
        post = re.sub(
            r'([😂🤪🤣😆😅😹😛😜😝🐸🐍😀😁😃😄😸😺🤭😇🤠😊🤗😎🌚🌝😉😋🤤🤩🥳🔥😌🤑💃💪✔✔️🤓🧐👍👌🙆‍♂🙆‍♀🙋‍♂🙋‍♀️✌️🤙👏🙌🤝'
            r'🙈🙉🙊☺️😘😗😙😚😽💋👄❤️❤️💘💝💖💗💓💞💕💌❣️🧡💛💚💙💜🖤♥💟😍🥰😻👨‍❤️‍💋‍👨👩‍❤️‍💋‍👩👨‍❤️‍👨👩‍❤️‍👩'
            r'💑💍👰🌸💮🏵🌹🥀🌺🌻🌼🌷💐😔😓😞😟☹️🙁😢😭😦😧😿😥😫😰😨🥺😩💔🤦‍♂️🤦‍♀️😪😴😷🤒🤕🤢🤮🤧😶🤐😏😼😈😒🤨😐😑😬🙂'
            r'🙃😕🤔🙄😖😵🥴🥵😱😳😮😯🙀😲😡😠🤬😤🤯😾👿😣🤥🤫💩🖕👊👎🙅‍♂️🙅‍♀️✌])\1{3,}',
            r'\1', post)
        post = re.sub('\؟+', ' ؟ ', post)
        post = re.sub('\!+', ' ! ', post)
        post = re.sub('\.+', ' . ', post)
        return post
    def remove_urls(post):
        return re.sub(r"http\S+"," ",post)
    def remove_hashtags(post):
        return re.sub(r"#\w+"," ",post )
    def remove_dirty_character(post):
        return re.sub(r"[-()\"#/@;:<>{}=~|.?,]"," ",post)
    from nltk import word_tokenize
    import pyarabic.arabrepr
    arepr = pyarabic.arabrepr.ArabicRepr()
    repr = arepr.repr

    from tashaphyne.stemming import ArabicLightStemmer
    ArListem = ArabicLightStemmer()
    word_list = "بعدسه سوريه ببساطه امطار غزيره في علي طريق حمص دمشق النبك واصوات الرعد في حمص تبشر بالخير القادم"

    # Define a function
    def steaming(word_list):
        
        wordsfilter=[]
        for a in word_tokenize(word_list):
            stem =  ArListem.light_stem(a)
            wordsfilter.append(stem)
        return " ".join(wordsfilter)
    def split_numbers_from_text(text):
        return " ".join(re.split(r'(\d+)', text)).strip()
    def remove_under_score(text):
        return re.sub(r"_"," ",text).strip()
    def remove_dirty(text):
        return re.sub(r"[^\w\s]","",text)
    post=remove_urls(post)
    post=remove_hashtags(post)
    post=delete_english_char(post)
    post=normaalize_write(post)
    post=remove_dirty_character(post)
#     post=steaming(post)
    post=split_numbers_from_text(post)
    post=remove_under_score(post)
    post=remove_dirty(post)

    return " ".join(re.split("\s+", post, flags=re.UNICODE))
```
