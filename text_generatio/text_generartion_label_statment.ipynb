```python
word_indices = dict((c, i) for i, c in enumerate(allword))
indices_word= dict((i, c) for i, c in enumerate(allword))
```

```python
def get_x_input_from_Sentenc(sentenc,step,maxlen):
    x=[]
    y=[]
    for i in range(0,len(sentenc)-maxlen,step):
    
        x.append(sentenc[i:i+maxlen])
        y.append(sentenc[i+maxlen])
    return x,y
def get_Xinput_from_all_data(sentences,step,maxlen):
    X=[]
    Y=[]
    for sentenc in sentences:
        x,y=get_x_input_from_Sentenc(sentenc.split(),step,maxlen)
        for xx in x:
            X.append(" ".join(xx))
        for yy in y:
            Y.append(yy)
    return X,Y
```

```python
X_train_string,Y_train_string=get_Xinput_from_all_data(data_set,1,maxlen)
```

```python
print('Vectorization...')

x = np.zeros((8000,maxlen, len(word_indices)), dtype=np.bool)
y = np.zeros((len(X_train_string[0:8000]), len(word_indices)), dtype=np.bool)
for i, sentence in enumerate(X_train_string[0:8000]):
    for t, char in enumerate(sentence.split(" ")):
        x[i, t, word_indices[char]] = 1    
    y[i, word_indices[Y_train_string[i]]] = 1
```

```python
def sample(preds, temperature=1.0):
    # helper function to sample an index from a probability array
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```
